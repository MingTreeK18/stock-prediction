```python
from sklearn.preprocessing import StandardScaler

df_copy = df.copy()
```

```python
df_copy['Date'] = pd.to_datetime(df_copy['Date'])
```

```python
# Tính số ngày kể từ ngày đầu tiên trong dữ liệu
df_copy['days_since_start'] = (df_copy['Date'] - df_copy['Date'].min()).dt.days

# Hiển thị DataFrame với cột 'days_since_start'
print(df_copy)
```

```python
df_copy=df_copy.drop(['Date'], axis = 1)
df_copy.head()
```

```python
# Tính MA50 và MA200
df_copy['MA50'] = df_copy['Close'].rolling(window=50).mean()
df_copy['MA200'] = df_copy['Close'].rolling(window=200).mean()

# Tính RSI (14 ngày mặc định)
def calculate_rsi(data, window=14):
    delta = data.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(window=window, min_periods=1).mean()
    avg_loss = loss.rolling(window=window, min_periods=1).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

df_copy['RSI'] = calculate_rsi(df['Close'])
df_copy.head()
```

```python
df_copy.tail()
```

```python

# Khởi tạo đối tượng StandardScaler
scaler = StandardScaler()

# Áp dụng chuẩn hóa cho DataFrame df
scaled_data = scaler.fit_transform(df_copy[['Open','High','Low','Close','Volume','MA50','MA200', 'RSI']])

# Chuyển đổi kết quả đã chuẩn hóa thành DataFrame và gán lại tên cột
scaled_df = pd.DataFrame(scaled_data, columns=['Open','High','Low','Close','Volume','MA50','MA200', 'RSI'])
first_col = df_copy[['days_since_start']]  # Lấy cột 'Days_Since_Start' ra
scaled_df.insert(0, 'days_since_start', first_col)
# In ra DataFrame đã chuẩn hóa
scaled_df.head()
```
